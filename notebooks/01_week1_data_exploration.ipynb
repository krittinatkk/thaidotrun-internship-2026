```python
# WEEK 1: FOUNDATION & SETUP COMPLETE!
# ThaiDotRun Data Analysis - First Notebook

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("üéâ WEEK 1 SETUP COMPLETE!")
print("=" * 50)
print(f"Python packages loaded successfully!")
print(f"Pandas version: {pd.__version__}")
print(f"NumPy version: {np.__version__}")

# Set up visualization
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("husl")
%matplotlib inline
```

```python
# Load the ThaiDotRun dataset
from pathlib import Path

data_path = Path("../data/raw/bkk_data_final.csv")
df = pd.read_csv(data_path)

print("üìä DATA LOADED SUCCESSFULLY!")
print(f"Dataset shape: {df.shape[0]} rows, {df.shape[1]} columns")
print("\nFirst 5 rows:")
df.head()
```

```python
# Basic information
print("üîç DATASET OVERVIEW")
print("=" * 50)
print(f"Total records: {len(df):,}")
print(f"Total columns: {len(df.columns)}")

# Column information
print("\nüìã COLUMNS:")
for i, col in enumerate(df.columns, 1):
    print(f"{i:2d}. {col}")

# Check data types
print("\nüìä DATA TYPES:")
print(df.dtypes.value_counts())

# Check for missing values
print("\nüîé MISSING VALUES:")
missing = df.isnull().sum()
missing_pct = (missing / len(df)) * 100
missing_df = pd.DataFrame({
    'Missing': missing,
    'Percentage': missing_pct.round(2)
})
print(missing_df[missing_df['Missing'] > 0])
```

```python
# 1. Event popularity
print("üèÉ MOST POPULAR EVENTS:")
top_events = df['eventName'].value_counts().head(10)
print(top_events)

# 2. Gender distribution
print("\nüë´ GENDER DISTRIBUTION:")
print(df['gender'].value_counts())

# 3. Ticket price analysis
print("\nüí∞ TICKET PRICE ANALYSIS:")
print(f"Min price: {df['ticketTypePrice'].min()} THB")
print(f"Max price: {df['ticketTypePrice'].max()} THB")
print(f"Average price: {df['ticketTypePrice'].mean():.2f} THB")

# 4. Virtual vs Physical events
print("\nüñ•Ô∏è VIRTUAL VS PHYSICAL EVENTS:")
print(df['isVirtual'].value_counts())
```
